**Sentiment Analysis**

```python
from transformers import pipeline
classifier=pipeline("sentiment-analysis")
classifier("good luck")
```

```python
classifier("This camera is not a bad choice for anyone who dislikes blurry photos.")
```

**Zero Shot Classification**

```python
classifier1=pipeline("zero-shot-classification")
classifier1(
    "csk loss this season but everyone taking ad-shooting to promote the product",
    candidate_labels=["education","sports","business","politics"]
)
```
