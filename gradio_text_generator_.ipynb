```python
import gradio as gr
from transformers import pipeline

print("Loading MiniCPM model... Please wait.")
generator = pipeline(
    "text-generation",
    model="openbmb/MiniCPM5-1B"
)
print("Model loaded successfully!")

# 2. Define the core logic function with fix parameters
def generate_poem(topic):
    if not topic.strip():
        return "Please enter a topic first!"

    # Structured chat-style prompt to guide small 1B models efficiently
    prompt = (
        f"User: Write a beautiful 4-line poem about {topic}.\n"
        f"Assistant: Here is the poem:\n"
    )

    # Optimized generation configuration to prevent looping and conversational filler
    result = generator(
        prompt,
        max_new_tokens=60,         # Kept tight to prevent rambling
        do_sample=True,
        temperature=0.85,          # Heightened for creative flair
        top_k=50,                  # Limits vocabulary pool to top quality choices
        top_p=0.92,                # Filters out incoherent tokens
        repetition_penalty=1.2,    # CRITICAL: Breaks the phrase looping bug
        return_full_text=False
    )

    poem = result[0]["generated_text"]

    # Post-processing cleanup to strip out any leaking structural tags
    cleanup_phrases = ["Answer:", "User:", "Assistant:", "No additional text.", "Here is the poem:"]
    for phrase in cleanup_phrases:
        poem = poem.replace(phrase, "")

    return poem.strip()

# 3. Build the UI Layout (Matching your wireframe structure)
with gr.Blocks(title="Poem Generator") as demo:
    # Heading matched to wireframe text
    gr.Markdown("# poem generator")

    with gr.Column(scale=1):
        # Input box matching your wireframe placeholder
        input_text = gr.Textbox(
            label="",
            placeholder="Write a poem on -----------------",
            lines=2
        )

        # Submit Button
        submit_btn = gr.Button("submit")

        # Output box to elegantly present the poem on the webpage
        output_text = gr.Textbox(
            label="Generated Poem",
            interactive=False,
            lines=5
        )

    # Link button trigger to core function
    submit_btn.click(
        fn=generate_poem,
        inputs=input_text,
        outputs=output_text
    )

# 4. Launch Web Server
if __name__ == "__main__":
    # Access via the URL printed in the terminal (usually http://127.0.0.1:7860)
    demo.launch()
```
